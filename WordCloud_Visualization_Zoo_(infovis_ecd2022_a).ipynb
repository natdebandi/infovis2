```python
# source: https://www.datacamp.com/community/tutorials/wordcloud-python

import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
# % matplotlib inline
```

```python
urls = ["https://biancabalzarini.github.io/infovis/s2/zoo.txt", 
"https://rfondato.github.io/infoviz/s2/zoo.txt", 
"https://jpcasal.github.io/infovis/s2/zoo.txt", 
"https://quarkzis.github.io/infovis/s2/zoo.txt", 
"https://kryari.github.io/infovis/s2/zoo.txt", 
"https://mdanielaraffom.github.io/infovis/s2/zoo.txt", 
"https://manueldvr.github.io/infovis/s2/zoo.txt", 
"https://natdebandi.github.io/infovis2/s2/zoo.txt", 
"https://hmreumann.github.io/ecd2022a/s2/zoo.txt"];

# urls = ['https://aaizemberg.github.io/infovis/data/zoo.txt']
```

```python
df = pd.DataFrame({'text': []})
for url in urls:
  print(url)
  df2 = pd.read_csv(url, header=None, names=['text'])
  df = df.append( df2.dropna(), ignore_index=True )
```

```python
df.tail()
```

```python
# armo el texto para el WordCloud
text = df.filter(items=['text']).squeeze().str.cat(sep=' ')

stopwords = { 'more', 'on', 'a', 'an', 'and', 'are', 'as', 'between', 'by', 'common', 'each', 'example', 
             'for', 'have', 'in', 'is', 'it', 'one', 'of', 'often', 'or', 'other', 'our', 'such', 
             'that', 'the', 'this', 'thus', 'to', 'tour', 'use', 'we', 'while', 'with' }

# Create and generate a word cloud image:

# wordcloud = WordCloud(max_words=200, width=800,height=600, stopwords=stopwords).generate(text)
wordcloud = WordCloud(max_words=200, width=800, height=600, stopwords=stopwords, background_color="black", collocations=True).generate(text)

# Display the generated image:
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# lo grabo como un .png
#
wordcloud.to_file("wordcloud_infovis-ecd-2022-a.png")
```
